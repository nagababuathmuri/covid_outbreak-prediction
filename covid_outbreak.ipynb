```python
#total cases in last five days
sns.set(rc={'figure.figsize':(3,3)})
sns.lineplot(x="date",y="total_cases",data=case)
plt.show()
```

```python
#Brazil case
brazil_case=covid[covid["location"]=="Brazil"]
brazil_case.head()
```

```python
#brazil for last 5 days
brazil_last_5_days=brazil_case.tail()
```

```python
#total cases in last five days
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=brazil_last_5_days)
plt.show()
```

```python
#understanding cases of brazil china and japan
brazil_china_japan=covid[(covid["location"]=="Brazil") | (covid["location"]=="China") | (covid["location"]=="Japan")]
```

```python
#plotting growth of cases across brazil china and japan
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="location",y="total_cases",data=brazil_china_japan,hue="date")
plt.show()
```

```python
#understanding cases of germany and spain
germany_spain=covid[(covid["location"]=="Germany") | (covid["location"]=="Spain")]
```

```python
#plotting growth of cases across germany and spain
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="location",y="total_cases",data=germany_spain,hue="date")
plt.show()
```

```python
#getting latest data
last_day_cases=covid[covid["date"]=="2020-05-24"]
last_day_cases
```
